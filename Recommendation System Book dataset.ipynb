```python
user_Books_df=Books_df.pivot_table(index='User.ID', columns='Book.Title', values='Book.Rating', aggfunc='sum')
```

```python
user_Books_df
```

```python
user_Books_df.index = Books_df["User.ID"].unique()
```

```python
user_Books_df.index
```

```python
#Impute those NaNs with 0 values
user_Books_df.fillna(0, inplace=True)
```

```python
user_Books_df
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim = 1 - pairwise_distances( user_Books_df.values,metric='cosine')
```

```python
user_sim
```

```python
#Store the results in a dataframe
user_sim_df = pd.DataFrame(user_sim)
user_sim_df
```

```python
#Set the index and column names to user ids 
user_sim_df.index = Books_df["User.ID"].unique()
user_sim_df.columns = Books_df["User.ID"].unique()
```

```python
user_sim_df.iloc[0:10, 0:10]  #SIMILARITY MATRIX
```

```python
np.fill_diagonal(user_sim, 0)
user_sim_df.iloc[0:5, 0:5]
```

```python
#IBCF
#Fetching ratings for Toy Story
Pleading_guilty_rating = user_Books_df["PLEADING GUILTY"] 
```

```python
Pleading_guilty_rating
```

```python
#Finding the correlation with different movies
similar_to_Pleading = user_Books_df.corrwith(Pleading_guilty_rating) 
```

```python
similar_to_Pleading
```

```python
corr_Pleading = pd.DataFrame(similar_to_Pleading, columns=['Correlation'])
corr_Pleading.dropna(inplace=True)
corr_Pleading.head()
```

```python
corr_Pleading[corr_Pleading['Correlation'] > 0].sort_values(by='Correlation', ascending=False).head(5)  
```
